```python
# shift operator --> shifts the bit by left or right
# left shift --> shifts the bits to the left by a specified number position filling 0's of the right
#  <<

5 << 2

#  0101 --> 010100
# 20
```

```python
#Right shift ---> remove the number of elements in the binary
# >>

20 >> 2

# 010100 --> 0101
# 5
```

```python
# Order of precedence --> execution of python statement will be from left to right
# ---> always parenthesis first
```
